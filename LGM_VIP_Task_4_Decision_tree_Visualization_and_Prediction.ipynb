```python
import numpy as npy
import pandas as pds
from sklearn.tree import DecisionTreeClassifier
```

# Reading Data

```python
dataframe_1 = pds.read_csv("Iris.csv")
dataframe_1
```

# Basic information about dataset

```python
dataframe_1.info()
```

```python
dataframe_1.describe()
```

```python
X_datasset = dataframe_1[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
X_datasset[0:5]
```

```python
Y_dataset = dataframe_1['Species']
Y_dataset[0:5]
```

# Importing decision tree classifier and splitting dataset in train and test dataset

```python
 from sklearn.model_selection import train_test_split
X_train_dataset, X_test_dataset, Y_train_dataset, Y_test_dataset = train_test_split(X_datasset,Y_dataset)
```

```python
decision_tree_for_species = DecisionTreeClassifier(criterion = 'entropy', max_depth = 4)
decision_tree_for_species
```

# Training the dataset 

```python
decision_tree_for_species.fit(X_train_dataset, Y_train_dataset)
```

# Predicting values on test dataset

```python
prediction_decision_tree = decision_tree_for_species.predict(X_test_dataset)
```

```python
print(prediction_decision_tree [0:5])
print(Y_test_dataset[0:5])
```

# Plotting the decison tree 

```python
from sklearn import metrics
import matplotlib.pyplot as plts
#print("DecisionTree Accuracy for predicting types of class: ",metrics.accuracy_score(Y_test_dataset, prediction_decision_tree)
print("DecisionTree Accuracy for predicting types of class: ",decision_tree_for_species.score(X_test_dataset, Y_test_dataset) * 100)
```

```python
import matplotlib.pyplot as plts
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
feature_names_iris = dataframe_1.columns[1:5]
class_names_iris= dataframe_1["Species"].unique().tolist()
decision_tree_for_species.fit(X_datasset, Y_dataset)
fig, axes = plts.subplots(nrows = 1,ncols = 1, figsize = (10,10), dpi = 375)

tree.plot_tree(decision_tree_for_species.fit(X_datasset, Y_dataset),feature_names = feature_names_iris, class_names = class_names_iris, filled = True);
```
